```python
serie_ipca_anual= "PRECOS_IPCAG"
serie_ipca_mensal = "PRECOS12_IPCA15G12"
pos_fix_anual = "('%s')" % serie_ipca_anual if serie_ipca_anual is not None else ""
pos_fix_mensal = "('%s')"% serie_ipca_mensal if serie_ipca_mensal is not None else ""
serie_anual_url = "http://www.ipeadata.gov.br/api/odata4/Metadados%s" % pos_fix_anual + '/Valores'
serie_mensal_url =  "http://www.ipeadata.gov.br/api/odata4/Metadados%s" % pos_fix_mensal+ '/Valores'
print(serie_anual_url)
print(serie_mensal_url)
```

```python
import requests
import pandas as pd
import json

ipca_mensal = requests.get(serie_mensal_url)
ipca_mensal_df = pd.DataFrame(json.loads(ipca_mensal.text)['value'])
ipca_mensal_df['VALDATA'] = pd.to_datetime(ipca_mensal_df['VALDATA'].apply(pd.to_datetime), utc=True)
ipca_mensal_df.head()
```

```python
ipca_anual = requests.get(serie_anual_url)
ipca_anual_df = pd.DataFrame(json.loads(ipca_anual.text)['value'])
ipca_anual_df['VALDATA'] = pd.to_datetime(ipca_anual_df['VALDATA'].apply(pd.to_datetime), utc=True)
ipca_anual_df.head()
```
